# Model quality improvment

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
from sklearn.neighbors import KNeighborsRegressor

from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

import matplotlib.pyplot as plt
import mplcyberpunk
from pylab import rcParams

from sklearn.decomposition import PCA
```

```python
from sklearn.datasets import load_boston

data = load_boston()
X = data.data
y = data.target
```

```python
for i in range(20):
    rand_state = np.random.randint(100) 
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=rand_state) 
    
    model = LinearRegression() 
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    r2 = r2_score(y_test, predictions) 
    mse = mean_squared_error(y_test, predictions)
    print('№{} Random state: {}; R2: {}; MSE: {}'.format(i, rand_state, round(r2, 5), round(mse, 5)))
```

```python
# for us best model is with best R2 score

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=85) 
model = LinearRegression() 
model.fit(X_train, y_train)
predictions = model.predict(X_test)
print('Best coef: ', model.coef_)
print('Best intercept: ', model.intercept_)
```

```python
rcParams['figure.figsize'] = 20, 10
```

```python
plt.style.use("cyberpunk")

plt.plot(predictions, marker='o')
plt.plot(y_test, marker='o')

mplcyberpunk.add_glow_effects()

plt.show()
```

**As we can see, the quality of our model is highly diversified. From poor 0.681 to reasonable 0.78**

LinearRegression has no super parametrs, so I will switch to SGD Regressor

```python
best = {'score': -100, 'penalty': '', 'loss': '', 'alpha': 0, 'epsilon': 0, 'model': 0}
loss = ['squared_loss', 'huber', 'epsilon_insensitive', 'squared_epsilon_insensitive']
penalty = ['l1', 'l2', 'elasticnet']

for p in penalty:
    for l in loss:
        for a in range(100):
            for e in range(10):
                model = SGDRegressor(penalty = p, 
                                     loss = l, 
                                     alpha = 0.0001 * a, 
                                     epsilon = 0.01 * e, 
                                     random_state = 42)
                results = cross_val_score(model, data.data, data.target, cv=10)
                avg_average = np.mean(results)
                
                #print('Loss function: {}\tPenalty: {}\tAlpha: {}\tEpsilon: {}\tAverage result: {}'.format(l, p,  0.0001 * a, 0.01 * e, round(avg_average, 3)))
                
                if avg_average > best['score']:
                    best['score'] = avg_average
                    best['penalty'] = p
                    best['loss'] = l
                    best['alpha'] = 0.0001 * a
                    best['epsilon'] = 0.01 * e
                    best['model'] = model
```

```python
best
```

```python
best['model'].fit(X_train, y_train)
predictions = best['model'].predict(X_test)
print('Best model R2 score: ', r2_score(y_test, predictions))
```

```python
best['model'].coef_
```

```python
best['model'].intercept_
```

Even with many attempts, the model quality is bad. Lets try to experiment with the best loss function - 'huber'.
Now I will change learing rate

```python
huber_best = {'score': -100, 'penalty': '', 'alpha': 0, 'model': 0}
learning_rate = ['constant', 'invscaling', 'adaptive']


for l in learning_rate:
    for e in range(1,1000):
        model = SGDRegressor(penalty = best['penalty'], 
                             loss = best['loss'], 
                             alpha = best['alpha'],
                             epsilon = best['epsilon'],
                             random_state = 42,
                             learning_rate = l,
                             eta0 = 0.0001 * e
                             )
        results = cross_val_score(model, X, y, cv=10)
        avg_average = np.mean(results)
                                
        if avg_average > best['score']:
            best['score'] = avg_average
            best['model'] = model
            best['learning_rate'] = l
            best['eta'] = 0.0001 * e
```

```python
best
```

```python
best['model'].fit(X_train, y_train)
predictions = best['model'].predict(X_test)
print('Best model R2 score: ', r2_score(y_test, predictions))
```

```python
plt.style.use("cyberpunk")

plt.plot(predictions, marker='o')
plt.plot(y_test, marker='o')

mplcyberpunk.add_glow_effects()

plt.show()
```

**still really week model**

# GridSearchCV 
I am not satisfied with SGDRegressor. I will switch to another model - Nearest Neighbors Regression.

Before I start to experiment with grid search, let's visualize the data using PCA to collapse features to one feature.

```python
pca = PCA(n_components=1)
pca.fit(X)
X_pca = pca.transform(X)
```

```python
plt.scatter(X_pca, y, marker='o')
mplcyberpunk.add_glow_effects()
plt.show()
```

Interesting pattern - we have blank space between ~100 and ~200. We see here some obvious groups.

```python
model = KNeighborsRegressor(n_neighbors=2)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
print('NNR first attempt R2 score: ', r2_score(y_test, predictions))
```

```python
plt.plot(predictions, marker='o')
plt.plot(y_test, marker='o')
mplcyberpunk.add_glow_effects()
plt.show()
```

<img src="https://sc.mogicons.com/share/not-bad-154.jpg">
The model gives much better results that SGD. But it could probably do better.

```python
k_range = list(range(1, 31))
algo = ['auto', 'ball_tree', 'kd_tree', 'brute']
param_grid = dict(n_neighbors=k_range, algorithm=algo)
```

```python
grid = GridSearchCV(KNeighborsRegressor(), 
                    param_grid=param_grid, 
                    cv=10, 
                    scoring='r2',
                    n_jobs=-1)
```

```python
grid.fit(X_train, y_train)
```

```python
# view the complete results (list of named tuples)
for k in grid.cv_results_:
    print(k, ":", grid.cv_results_[k][0])
```

```python
# create a list of the mean scores only
test_scores = grid.cv_results_['mean_test_score']
print(test_scores)
```

```python
# examine the best model

# Single best score achieved across all params (k)
print(grid.best_score_)

# Dictionary containing the parameters (k) used to generate that score
print(grid.best_params_)

# Actual model object fit with those best parameters
# Shows default parameters that we did not specify
print(grid.best_estimator_)
```

```python
model = grid.best_estimator_
model.fit(X_train, y_train)
predictions = model.predict(X_test)
plt.plot(predictions, marker='o')
plt.plot(y_test, marker='o')
mplcyberpunk.add_glow_effects()
plt.show()
```

# Final thoughts

As we can see, a selection of a model and parameters is very important. I got 0.5 R2 score with no very suitable model (neighbors are more about classification). Even with my simple assumption that there are 2 clusters, I got better results. But {'algorithm': 'auto', 'n_neighbors': 4} should be better on real data. We can check this by our first method.

```python
for i in range(20):
    rand_state = np.random.randint(100) 
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=rand_state) 
    
    model = KNeighborsRegressor(n_neighbors=2)
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    r2 = r2_score(y_test, predictions) 
    mse = mean_squared_error(y_test, predictions)
    print('№{} Random state: {}; R2: {}; MSE: {}'.format(i, rand_state, round(r2, 5), round(mse, 5)))
```
