```python
import pandas as pd
import pickle

geo = pd.read_excel('geo_comment.xlsx')
comment = geo['comment']
```

# Document-term matrix

```python
from sklearn.feature_extraction.text import HashingVectorizer, CountVectorizer
# we must remove stop words
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

sw = stopwords.words('russian')
with open('comment_stop_words.txt', encoding='utf-8') as f:
    our_stop_words = f.read().splitlines()
    f.close()

sw = sw + our_stop_words

# We will remove digits
word_pattern = r'[a-zA-Zа-яА-Я]+'

cv = CountVectorizer(stop_words= sw, token_pattern=word_pattern)
data_cv = cv.fit_transform(comment)

data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())

del cv
del data_cv
```

# Save document term matrix to file
Need a lot of memory for that

```python
f = open('dtm.obj', 'wb')
pickle.dump(data_dtm, f)
f.close()
```

# Load document term matrix

```python
f = open('dtm.obj', 'wb')
data_dtm = pickle.load(f)
f.close()
```

```python
print('Text count: ', data_dtm.shape[0], ' Word count: ', data_dtm.shape[1])
```

# Word stats

### Yandex Mystem

```python
from pymystem3 import Mystem


def lemma(word):    
    m = Mystem()
    return m.lemmatize(word)[0]
```

### Morphy

```python
import pymorphy2


def lemma(word):
    morph = pymorphy2.MorphAnalyzer()
    p = morph.parse(word)[0]
    return p.normal_form
```

```python
%%time

words = {}

for word in data_dtm.columns:
    # first lemmatize colum
    w = lemma(word)
    
    if w in words:
        words[w] += data_dtm[word].sum()
    else:
        words[w] = data_dtm[word].sum() 

# save to DF, sort
w = pd.DataFrame.from_dict(words, orient='index')
w.sort_values(by=0,ascending=False, inplace=True)
```

```python
# save to csv
w.to_csv('word_stats.csv', encoding='utf-8')

# f = open('words.obj', 'wb')
# pickle.dump(words, f)
# f.close()
```

```python
from matplotlib import pyplot as plt
from wordcloud import WordCloud
wc = WordCloud().generate_from_frequencies(words)

plt.figure(figsize=(15,12))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
#fig.set_facecolor("white")
plt.show()
```

# Lemmatization

```python
from pymystem3 import Mystem
m = Mystem()

# dictionary of only lemma columns
dc = {}
i = 0 
l = len(data_dtm.columns)

for column in data_dtm.columns:
    i += 1
    
    lemma_column = m.lemmatize(column)[0]
   
    if lemma_column in dc:
        print(f'Duplicate column #{i}')
            
        #sum words count suitable for one lemma
        print('Sum was ', dc[lemma_column].sum())
        dc[lemma_column] = dc[lemma_column] + data_dtm[column]
        print('Sum now ', dc[lemma_column].sum())
    else:
        # init lemma column
        dc[lemma_column] = data_dtm[column]
    
    print(f'Lemmatize of {column} to {lemma_column} done. Step {i} of {l}')

    
f = open('lemmatize.obj', 'rb')
pickle.dump(f, dc)
f.close()
```
