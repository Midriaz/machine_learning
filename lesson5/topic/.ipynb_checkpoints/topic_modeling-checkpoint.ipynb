```python
geo = pd.read_excel('geo_comment.xlsx')
comment = geo['comment']
```

# Document-term matrix

```python
# we must remove stop words
from nltk.corpus import stopwords

sw = stopwords.words('russian')
with open('comment_stop_words.txt', encoding='utf-8') as f:
    our_stop_words = f.read().splitlines()
    f.close()

sw = sw + our_stop_words
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# We will remove digits
word_pattern = r'[a-zA-Zа-яА-Я]+'

cv = CountVectorizer(stop_words= sw, token_pattern=word_pattern)
data_cv = cv.fit_transform(comment)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
del(cv)
del(data_cv)

data_dtm
```

# Lemmatization

```python
from pymystem3 import Mystem
m = Mystem()

# dictionary of only lemma columns
dc = {}

for column in data_dtm.columns:
    lemma_column = m.lemmatize(column)[0]
   
    if lemma_column in dc:
        #sum words count suitable for one lemma
        dc[lemma_column] += data_dtm[column]
    else:
        # init lemma column
        dc[lemma_column] = data_dtm[column]
```
